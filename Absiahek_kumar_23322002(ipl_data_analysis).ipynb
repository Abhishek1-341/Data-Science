```python
# top tan batsman with highest run rate
#conditions--> player should have played at least 50 balls
mask=deliveries.groupby('batter')['batsman_runs'].count()>60
a=deliveries.groupby('batter')['batsman_runs'].count()[mask]
batsmans=list(a.index)

mask=deliveries['batter'].isin(batsmans)
data=deliveries[mask]
runs=data.groupby('batter')['batsman_runs'].sum()
balls=data.groupby('batter')['batsman_runs'].count()
run_rate=(runs*100)/balls

run_rate.sort_values(ascending=False)[:20]
```

```python
run_rate.sort_values(ascending=False)[:10].plot(kind='bar',color=l)
```

```python
# bowle with lowest economy rate
#condition---> bowle have bowled atleast 300 balls
mask=deliveries['bowler'].value_counts()>300
a=deliveries['bowler'].value_counts()[mask]
bowlers=list(a.index)

mask=deliveries['bowler'].isin(bowlers)
data=deliveries[mask]
runs=data.groupby('bowler')['total_runs'].sum()
balls=data.groupby('bowler')['total_runs'].count()

economy_rate=(runs*6)/balls
economy_rate.sort_values()[:10]
```

```python
economy_rate.sort_values()[:10].plot(kind='bar',color=l)
```

```python
# analysing the relationship between total no of 4`s and total no of 6`s by players
#condition player have hit atleast 50 6's and 50 4's 
mask=( deliveries['batsman_runs']==4 ) 
a=deliveries[mask]


mask=( deliveries['batsman_runs']==6) 
b=deliveries[mask]

a=a.groupby('batter')['batsman_runs'].count()
b=b.groupby('batter')['batsman_runs'].count()
print(len(a),len(b))
```

```python
data = pd.merge(a, b, left_index=True, right_index=True, how='inner', suffixes=["total_4's", "total_6's"])
mask=( data["batsman_runstotal_4's"]>50 )  |( data["batsman_runstotal_6's"]>50)
data=data[mask]

sns.scatterplot(x=data["batsman_runstotal_4's"], y=data["batsman_runstotal_6's"])
plt.title("relationship between total no of 4's and total no of 6's".title(),color="#32CD32")
plt.show()
```

```python
sns.regplot(x=data["batsman_runstotal_4's"], y=data["batsman_runstotal_6's"], line_kws={'color':'#32CD11'})
plt.title("relationship between total no of 4's and total no of 6's".title(),color="#32CD32")
plt.show()
```

```python
np.corrcoef(np.array(data["batsman_runstotal_4's"]),np.array(data["batsman_runstotal_6's"]))
# so there a significant  relationship between total no of sixes and total no. of fours hit by players
# generally player with more no of tatal sixes also have more no of total fours
```

```python
# relations between runs verses overs
mask=deliveries['inning']==1
a=deliveries[mask]

b=a.groupby(['match_id','over'])["total_runs"].sum()
full_first_inning=list( b.reset_index()['match_id'].value_counts()[b.reset_index()['match_id'].value_counts()>19].index )

```

```python
mask=b.reset_index()['match_id'].isin(full_first_inning)
data=b.reset_index()[mask]
# so we are only examining first ining and only those matches which are played till last over
avg=data.groupby('over')['total_runs'].sum()/len(data.groupby('match_id'))
avg
```

```python
sns.catplot(x=data['over']+1, y=data['total_runs'], kind='box', height=10, aspect=2,color="#aaeeaa")
plt.xlabel('Overs',size=30)
plt.ylabel('Runs per Over',size=30)
plt.xticks(size=20)
plt.yticks(np.arange(0,38,2),size=20)
plt.title('Runs per over for first Ining'.title(),size=40,color='#567888')
plt.grid()
```

```python

sns.catplot(x=data['over']+1, y=data['total_runs'], kind='violin', height=10, aspect=2,color="#eeaaaa")
plt.xlabel('Overs',size=30)
plt.ylabel('Runs per Over',size=30)
plt.xticks(size=20)
plt.yticks(np.arange(0,38,2),size=20)
plt.title('Runs per over for first Ining'.title(),size=40,color='#567888')
plt.grid()
```

```python
sns.catplot(kind='point',x=avg.index+1, y=avg.values, height=10, aspect=2,color="#4444ff")
#sns.catplot(x=data['over'], y=data['total_runs'], kind='strip', height=10, aspect=2,color="#eeaaaa")
plt.xlabel('Overs',size=30)
plt.ylabel('Runs per Over',size=30)
plt.xticks(size=20)
plt.yticks(np.arange(3,13,1),size=20)
plt.title('Runs per over for first Ining'.title(),size=40,color='#567888')
plt.grid()

# as we see in graph that firstly average runs increases till 6th over then there is a sudden fall then afer that runs per over increases till end
#while moving from 6th to 7th over run decreases because first 6 overs are power play
# During the powerplay overs, only two fielders are allowed outside the 30-yard circle
```

```python
# relations between wickets verses overs
mask=deliveries['inning']==1
a=deliveries[mask]
data=a.groupby(['match_id','over'])['is_wicket'].sum().reset_index()
avg2=data.groupby('over')['is_wicket'].sum()/len(data.groupby('match_id'))
avg2
```

```python
sns.catplot(kind='point',x=avg2.index+1, y=avg2.values, height=10, aspect=2,color="#44ff44")
plt.xlabel('Overs',size=30)
plt.ylabel('Wickets per Over',size=30)
plt.xticks(size=20)
plt.yticks(np.arange(0,1,0.1),size=20)
plt.title('wickets per over for first Ining'.title(),size=40,color='#567888')
plt.grid()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(x=avg.index,y=avg.values,color="#4444ff",label="Runs per Over")
sns.pointplot(x=avg2.index,y=avg2.values,color="#44ff44",label="Wickets per Over")

plt.xlabel('Overs',size=20)
plt.ylabel('Runs/Wickets per Over',size=20)
plt.xticks(size=20)
plt.yticks(np.arange(0,30,2),size=20)
plt.legend(title='Metric', loc='upper left', fontsize=20, title_fontsize='20')
plt.show()
# we need to scale up and down so we could analys trends
```

```python
a=list(avg.values)
a=sum(a)/len(a)
b=(list(avg2.values))
b=sum(b)/len(b)
avg=avg/a
avg2=avg2/b
# scalling down/up so data varry only in between 0 and 1
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(x=avg.index,y=avg.values,color="#4444ff",label="Runs per Over")
sns.pointplot(x=avg2.index,y=avg2.values,color="#44ff44",label="Wickets per Over")

plt.xlabel('Overs',size=20)
plt.ylabel('Runs/Wickets per Over',size=20)
plt.xticks(size=20)
plt.yticks(np.arange(0,2,0.2),size=20)
plt.legend(title='Metric', loc='upper left', fontsize=20, title_fontsize='20')
plt.grid()
plt.show()
```

```python
np.corrcoef(avg.values,avg2.values)
# so we are having significant correlation
```

```python
np.corrcoef(avg.values[:-1],avg2.values[:-1])
# if we exclude last ovwe correlation further increases
```

```python
# Variation of no of sixes with over and team for first batting team who have played till 20th over
mask=(deliveries['inning']==1) & (deliveries['batsman_runs']==6)
a=deliveries[mask]
data=a.groupby(['match_id','over'])['batsman_runs'].count().reset_index()


data=pd.merge(data,matches , how='inner'  , left_on='match_id',right_on='id')
data['over']+=1

data.pivot_table(index='team1',columns='over',values='batsman_runs',aggfunc='sum')
```

```python
# for some specific teams only
mask=data['team1'].isin([  "Mumbai Indians", "Kolkata Knight Riders", "Royal Challengers Bangalore", "Chennai Super Kings", "Rajasthan Royals"  ,"Kings XI Punjab" ,"Sunrisers Hyderabad"   ])
data=data[mask]
p=data.pivot_table(index='team1',columns='over',values='batsman_runs',aggfunc='sum')
p
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(p, cbar=False,linewidths=0.5,annot=True,fmt='d',annot_kws={"size": 20})
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel("Overs",size=30)
plt.ylabel("team1",size=30)
plt.title('Variation of no of sixes with over and team ',size=25,color="#ee4444")
plt.show()
```

```python
#Variation of no of fours with over and team for first batting team who have played till 20th over
mask=(deliveries['inning']==1) & (deliveries['batsman_runs']==4)
a=deliveries[mask]
data=a.groupby(['match_id','over'])['batsman_runs'].count().reset_index()

data=pd.merge(data,matches , how='inner'  , left_on='match_id',right_on='id')
data['over']+=1

mask=data['team1'].isin([  "Mumbai Indians", "Kolkata Knight Riders", "Royal Challengers Bangalore", "Chennai Super Kings", "Rajasthan Royals"  ,"Kings XI Punjab" ,"Sunrisers Hyderabad"   ])
data=data[mask]
p=data.pivot_table(index='team1',columns='over',values='batsman_runs',aggfunc='sum')

plt.figure(figsize=(20,10))
sns.heatmap(p, cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='winter',annot_kws={"size": 20,'color':'#000000'})
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel("Overs",size=30)
plt.ylabel("team1",size=30)
plt.title('Variation of no of fours with over and team '.title(),size=25,color="#4444ee")
plt.show()
```
